```python
# Extract characters 16 to 19 from the CellID index
adata.obs['plate'] = adata.obs.index.str[15:19]

adata.obs['plate'].value_counts()
```

```python
#0101,0201,0301,0401 is the first 

import numpy as np

# Define conditions and corresponding batch values
conditions = [
    adata.obs['plate'].isin(['0101', '0201', '0301', '0401']),
    adata.obs['plate'].isin(['1401', '1402', '1403'])
]
batches = [1, 2]

# Use np.select to assign batch values based on the conditions
adata.obs['batch'] = np.select(conditions, batches, default=3)

adata.obs
```

```python
adata.obs['time'] = 'time'
adata.obs.loc[adata.obs_names.str.contains('0101|0601'), 'time'] = '0h'
adata.obs.loc[adata.obs_names.str.contains('0701'), 'time'] = '48h'
adata.obs.loc[adata.obs_names.str.contains('0201|0801'), 'time'] = '72h'
adata.obs.loc[adata.obs_names.str.contains('0301|0302|0901'), 'time'] = '96h'

#whatever is left go to 120h
adata.obs.loc[adata.obs['time'] == 'time', 'time'] = '120h'
```

```python
# Assuming you've already run a clustering method and have an embedding (like UMAP or tSNE)
scv.tl.velocity_embedding(adata, basis='X_pca')
scv.pl.velocity_embedding(adata, basis='X_pca', arrow_length=3, arrow_size=2, dpi=120)
```

```python
adata.obs['time'].value_counts()
```

```python
sc.pp.filter_cells(adata, min_genes=3000)
sc.pp.filter_genes(adata, min_cells=3)  
adata.var['mt'] = adata.var_names.str.startswith('MT-')  # annotate the group of mitochondrial genes as 'mt'
sc.pp.calculate_qc_metrics(adata, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
adata.var['mt'] = adata.var_names.str.startswith('MT-')  # annotate the group of mitochondrial genes as 'mt'
sc.pp.calculate_qc_metrics(adata, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
sc.pl.scatter(adata, x='total_counts', y='pct_counts_mt')
sc.pl.scatter(adata, x='total_counts', y='n_genes_by_counts')

sc.pp.normalize_total(adata, target_sum=1e4)

sc.pp.log1p(adata)

sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)

sc.pl.highly_variable_genes(adata)

adata = adata[:, adata.var.highly_variable]


sc.pp.scale(adata, max_value=10)

sc.pp.combat(adata, key='batch')

sc.tl.pca(adata, svd_solver='arpack')

sc.pl.pca(adata, color=['time'])

sc.pp.neighbors(adata, n_neighbors=20, n_pcs=20)
sc.tl.draw_graph(adata)


sc.tl.leiden(adata, resolution=0.3)



#umap plot
sc.tl.umap(adata)
sc.pl.umap(adata, color=['time', 'leiden','plate'])


```
